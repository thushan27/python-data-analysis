```python
link = 'https://en.wikipedia.org/wiki/List_of_all-time_NFL_win%E2%80%93loss_records'
html_content = pd.read_html(link)
new_nfl_frame = html_content[1]
```

```python
"""
using the isin() function which will return True/False for each value and pandas will
get these specific values from the division column. can assign each condition as a variable to make it simpler. 
"""
sub1 = new_nfl_frame[(new_nfl_frame['Won']>500) & (new_nfl_frame['Division'].isin(['AFC South','NFC North','NFC East']))]
sub2 = new_nfl_frame[(new_nfl_frame['Lost']<500) & (new_nfl_frame['Division'].isin(['AFC South','NFC North','NFC East']))]
```

```python
sub1
```

```python
sub2
```

```python
'''
takes data frame sub2 and places it under data frame sub1 to form a new data frame
'''
pd.concat([sub1,sub2])
```

```python
"""
this is merging 2 data frames. NaN appears as pandas reads the index number and the corresponding values in that row.
"""
merged = pd.concat([sub1,sub2], axis = 1)
```

```python
"""
.dropna() drops all rows with any NaN
.dropna(how = 'all') drops rows with all entries = NaN
.dropna(thresh = n) drops rows with n entries = NaN
.fillna(2) fills all NaN entries with int 2
.fillna({0:0,1:1}) fills 1st column with 0s, 2nd column with 1s
"""
merged.dropna()
```
